## Important Concepts

### probability mass function and probability density function

```python
import random
import numpy as np
```

Generate all possible sum of outcomes

```python
dice = [1,2,3,4,5,6]
probs = [1/6,1/6,1/6,1/6,1/6,1/6]
sums = sorted(np.unique([dice[i] + dice[j] for i in range(6) for j in range(6)]))
```

Calculate all associated probabilities

```python
from collections import OrderedDict
```

```python
res = OrderedDict()
for s in sums:
    res[s] = 0
    for i in range(6):
        for j in range(6):
            if dice[i]+dice[j]==s:
                res[s] += probs[i]*probs[j]             
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(8,6))
plt.rcParams.update({'font.size': 22})
plt.bar(res.keys(),res.values())
plt.title("Probabilities of Two Dice Sum")
plt.xlabel("Sum Value")
plt.ylabel("Probability");
```

## Common discrete probability distributions

Bernoulli distribution

```python
p = 0.7
samples = [random.random() < 0.7 for _ in range(1000)]
print(np.mean(samples),np.var(samples))
```

Binomial distribution

```python
p = 0.8
X = [i for i in range(1,101)]
n = 100
Fx = [np.math.factorial(n)/(np.math.factorial(n-k)*np.math.factorial(k))*p**k*(1-p)**(n-k) for k in X]
random.seed(2020)
K = []
for trial in range(1000):
    k = np.sum([random.random() < p for _ in range(n)])
    K.append(k)
plt.figure(figsize=(8,6))
plt.hist(K,bins=30,density=True,label="Computational Experiment PMF");
plt.plot(X,Fx,color="r",label="Theoretical PMF",linestyle="--")
plt.legend();
```

Poisson distribution

```python
lambdas = [2,4,6,16]
K = [k for k in range(30)]
plt.figure(figsize=(8,6))
for i,l in enumerate(lambdas):
    plt.plot(K,[np.exp(-l)*l**k/np.math.factorial(k) for k in K],
             label=str(l),
            marker=i*2)
plt.legend()
plt.ylabel("Probability")
plt.xlabel("Values")
plt.title("Parameterized Poisson Distributions");
```

Exponential distribution

```python
lambdas = [0.2,0.4,0.8,1.0]
K = [0.5*k for k in range(15)]
plt.figure(figsize=(8,6))
for i,l in enumerate(lambdas):
    plt.plot(K,[np.exp(-l*k)*l for k in K],
             label=str(l),
            marker=i*2)
plt.legend()
plt.ylabel("Probability")
plt.xlabel("Values")
plt.title("Parameterized Exponential Distributions");
```

normal distribution with different expectations

```python
mus = [-1,0,1]
K = [0.2*k-5 for k in range(50)]
plt.figure(figsize=(8,6))
for i,mu in enumerate(mus):
    plt.plot(K, [1/(np.sqrt(2*np.pi))*np.exp(-(k-mu)**2/2) for k in K],
             label=str(mu),
            marker=i*2)
plt.legend()
plt.ylabel("Probability")
plt.xlabel("Values")
plt.title("Parameterized Normal Distributions");
```
